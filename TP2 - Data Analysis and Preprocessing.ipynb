```python
import wfdb
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn import decomposition
from scipy.signal import butter, lfilter, freqz, medfilt
import sklearn.neural_network as snn
from sklearn.metrics import confusion_matrix
import itertools
import copy
from IPython.display import display
from scipy.fftpack import fft, fftfreq
```

```python
print('Fetching data...')
signals2, fields2=wfdb.srdsamp('14172', channels=[0,1], pbdir = 'ltdb') 
annotation = wfdb.rdann('14172', 'atr', pbdir = 'ltdb')
print('Data fetched!')
```

# Preparación de la señal

Se descartan picos de latidos tipo 'F' y 'a' además de los puntos '~'.

También se descartan los primeros 5 minutos de señal así como también el último latido.

```python
#Me quedo con los picos de cada ECG que es donde están las anotaciones
picos=copy.copy(annotation.annsamp)
clasificacion=copy.copy(annotation.anntype)

#Transpongo las señales leídas para poder operar mas comodamente
signals=np.matrix.transpose(copy.copy(signals2))

#Descarto picos que no me interesan
fs = 128
tam=64 #Tamaño del latido que tomo a la izquierda y a la derecha de la anotación
threshold = 5*60*fs #los picos iniciales menores a 5 minutos se descartan

for j in range(len(picos))[::-1]:
    if clasificacion[j] in ['~', 'F', 'a']:
        del clasificacion[j]
        picos = np.delete(picos, j)
        
L = 0
left_padded_picos = picos[picos <= tam + threshold]
left_padded_picos = left_padded_picos[-L:]
picos[-1]=0 #descarto el último pico por si el latido está interrumpido
picos = picos[picos > tam + threshold]
clasificacion = clasificacion[len(clasificacion)-len(picos)-1:len(clasificacion)-1]
        
```

```python
inter_beat_interval = []
inter_beat_interval.extend(left_padded_picos)
inter_beat_interval.extend(picos)
```

# Preprocesamiento - Filtrado en frecuencia

```python
plt.rcParams['figure.figsize'] = (20.0, 12.0)
T = 1.0 / fs
N = len(signals[0])
xf = fftfreq(signals[0].size, T)
fig, ax = plt.subplots(nrows=2,ncols=2)

### Sin filtrar ###
# Channel 0
plt.subplot(2,2,1)
plt.title("Channel 0")
plt.ylabel('Original')
yf = fft(signals[0])
plt.loglog(xf[:N//2], 2.0/N * np.abs(yf[:N//2]))

# Channel 1
plt.subplot(2,2,2)
plt.title("Channel 1")
yf = fft(signals[1])
plt.loglog(xf[:N//2], 2.0/N * np.abs(yf[:N//2]))

### FILTRADO ###
order = 8
cutoff = 0.5 # Hz
b, a = butter(order, cutoff / (fs / 2), btype='high')
signals[0] = lfilter(b, a, signals[0])
signals[1] = lfilter(b, a, signals[1])

order = 12
cutoff = 35 # Hz
b, a = butter(order, cutoff / (fs / 2), btype='low')
signals[0] = lfilter(b, a, signals[0])
signals[1] = lfilter(b, a, signals[1])
################

### Filtradas ###
# Channel 0
plt.subplot(2,2,3)
plt.title("Channel 0")
plt.ylabel('Filtradas')
plt.xlabel('Frecuencia (Hz)')
yf = fft(signals[0])
plt.loglog(xf[:N//2], 2.0/N * np.abs(yf[:N//2]))

# Channel 1
plt.subplot(2,2,4)
plt.title("Channel 1")
plt.xlabel('Frecuencia (Hz)')
yf = fft(signals[1])
plt.loglog(xf[:N//2], 2.0/N * np.abs(yf[:N//2]))

plt.show()
```

La adición de un filtro pasa-altos al preprocesamiento puede mejorar considerablemente la detección de latidos prematuros nodales y supraventriculares. Esto es apreciable ya que se alcanzan resultados similares al reducir la cantidad de componentes principales usados y, si se mantienen constantes, aumenta la tasa de verdaderos positivos alrededor de 10 puntos porcentuales para ambos casos.



# Análisis de los distintos tipos de latidos

```python
latidos=np.zeros([len(picos),4*tam + L]) #Reservo espacio para la matriz de datos cantLatidos x 2*tam
for i in range(len(picos)):
    channel0 = signals[0,picos[i]-tam:picos[i]+tam]
    channel1 = signals[1,picos[i]-tam:picos[i]+tam]
    latidos[i]= np.concatenate((channel0, channel1, np.array(inter_beat_interval[i:(i+L)])), axis=0) #tomo tam muestras a la izquerda y a la derecha de la anotación
```

```python
### Select training and testing samples ###
# Separate heartbeats by classification
picos_normales = []
picos_v_prematuros = []
picos_sv_prematuros = []
picos_n_prematuros = []
for j in range(len(picos)):
    if clasificacion[j] == 'J':
        picos_sv_prematuros.append(latidos[j])
    elif clasificacion[j] == 'N':
        picos_normales.append(latidos[j])
    elif clasificacion[j] == 'S':
        picos_n_prematuros.append(latidos[j])
    elif clasificacion[j] == 'V':
        picos_v_prematuros.append(latidos[j])
        
picos_normales = np.array(picos_normales)
picos_v_prematuros = np.array(picos_v_prematuros)
picos_sv_prematuros = np.array(picos_sv_prematuros)
picos_n_prematuros = np.array(picos_n_prematuros)

# A continuación calculo la media y el desvio de cada elemento del vector para cada clasificacion.
medias_n_prematuros = picos_n_prematuros.mean(axis=0)
desv_n_prematuros = picos_n_prematuros.std(axis=0)

medias_v_prematuros = picos_v_prematuros.mean(axis=0)
desv_v_prematuros = picos_v_prematuros.std(axis=0)

medias_sv_prematuros = picos_sv_prematuros.mean(axis=0)
desv_sv_prematuros = picos_sv_prematuros.std(axis=0)

medias_normales = picos_normales.mean(axis=0)
desv_normales = picos_normales.std(axis=0)

```
